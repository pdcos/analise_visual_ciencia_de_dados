#### (c) Sintetize com base em https://betterstack.com/community/guides/scaling-python/python-errors/ os erros comuns em Python

- **SyntaxError:** Este erro ocorre quando o usuário escreve algo que não vai de acordo com as regras de sintaxe do interpretador Python.

- **IndentationError:** É causado pela presença (ou falta) de um espaçamento correto entre blocos de código.

- **NameError:** Ocorre quando uma variável que não foi declarada é chamada ao longo da execução. Erros de digitação são uma causa bem comum para isso.

- **ValueError:** Ocorre quando uma função ou método recebem uma variável com tipagem diferente da esperada.

- **UnboundLocalError:** Acontece quando uma variável é declarada localmente dentro de uma função mas é acessada antes da declaração.

- **TypeError:** É causado pela tentativa de realização de uma operação que um determinado objeto não suporta.

- **UnicodeError:** Ocorre por problemas/conflitos de encodificação e/ou decodificação de strings.

- **ZeroDivisionError:** É causado quando ocorre uma divisão por zero.

- **FileNotFoundError:** Este acontece quando o script tenta fazer uma operação em um arquivo inexistente.

- **ModuleNotFoundError:** É ocasionado através da importação de módulos/bibliotecas que não existem ou não foram instaladas localmente.

- **MemoryError:** Ocorre quando o sistema fica sem memoria ao longo da execução de um script.

- **PermissionError:** É ocasionado através da execução de operações que requerem um privilégio. Como exemplo disso, temos o acesso a determinados diretórios essenciais para o funcionamento do sistema operacional, que requerem acesso de administrador para edição de arquivos.

- **IndexError:** Ocorre quando se tenta acessar uma posição (ou índice) inexistente em uma lista, tupla ou string.

- **KeyError:** É semelhante ao *IndexError*, mas referente a dicionários. Aqui o erro acontece quando se fornece uma chave inexistente a um dicionário.

- **AttributeError:** Ocorre quando se tenta acessar um atributo (parâmetro ou método) inexistente dentro de uma classe.
